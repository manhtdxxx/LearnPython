```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
df['age_scaled'] = df.age / 100
```

```python
X = df[['age_scaled','affordibility']]
Y = df.bought_insurance
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, random_state=25)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=5000)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp_test_proba = model.predict(X_test)
yp_test_proba
```

```python
yp_test = np.where(yp_test_proba > 0.5, 1, 0)
yp_test.reshape(1,-1)
```

```python
y_test.values
```

```python
coef, intercept = model.get_weights()
coef, intercept
```

```python
def prediction(x1,x2):
    weighted_sum = coef[0]*x1 + coef[1]*x2 + intercept
    return 1 / (1 + np.exp(-weighted_sum))
```

```python
prediction(X_test.values[:,0], X_test.values[:,1])
```

### From scratch

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def binary_crossentropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = [max(i,epsilon) for i in y_pred]
    y_pred = [min(i,1-epsilon) for i in y_pred]
    y_pred = np.array(y_pred)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
def gradient_descent(x1, x2, y_true, epochs):
    w1 = w2 = 1
    bias = 0
    learning_rate = 0.5
    
    for i in range (epochs):
        # activation function
        weighted_sum = w1*x1 + w2*x2 + bias
        y_pred = sigmoid(weighted_sum)
        
        # loss function
        loss = binary_crossentropy(y_true, y_pred)
        
        #
        print('epochs={}, w1={}, w2={}, bias={}, loss={}'.format(i, w1, w2, bias,loss))

        # derivatives of loss function wih respect to weights,bias
        d_w1 = -np.mean((
            (y_true / y_pred) - ((1-y_true) / (1-y_pred)) # o hm ca binary_crossentropy theo y_pred
        ) * (np.exp(-weighted_sum) / (1+np.exp(-weighted_sum))**2 # o hm ca y_pred theo weighted_sum
            ) * x1) # o hm ca weighted_sum theo w1
        
        d_w2 = -np.mean((
            (y_true / y_pred) - ((1-y_true) / (1-y_pred)) # o hm ca binary_crossentropy theo y_pred
        ) * (np.exp(-weighted_sum) / (1+np.exp(-weighted_sum))**2 # o hm ca y_pred theo weighted_sum
            ) * x2) # o hm ca weighted_sum theo w2
        
        d_bias = -np.mean((
            (y_true / y_pred) - ((1-y_true) / (1-y_pred)) # o hm ca binary_crossentropy theo y_pred
        ) * (np.exp(-weighted_sum) / (1+np.exp(-weighted_sum))**2 # o hm ca y_pred theo weighted_sum
            ) * 1) # o hm ca weighted_sum theo bias
        
        # adjust weights,bias
        w1 = w1 - learning_rate * d_w1
        w2 = w2 - learning_rate * d_w2
        bias = bias - learning_rate * d_bias

    #
    return w1,w2,bias,loss
```

```python
print(gradient_descent(X_train.values[:,0], X_train.values[:,1], y_train, 5000))
```
